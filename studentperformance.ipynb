```python
print(df["gender"].unique())
print(df["race/ethnicity"].unique())
print(df["parental level of education"].unique())
print(df["lunch"].unique())
print(df["test preparation course"].unique())
```

```python
df['total']=df['math score']+df['reading score']+df['writing score']
df['average']=(df['total'])/3
df.head()
```

```python
print(f"no. of students who have scored more than 90 in any one of the subjects: {df[(df['math score'] > 90) | (df['reading score'] > 90) | (df['writing score'] > 90)].shape[0]}")
df[(df['math score'] > 90) | (df['reading score'] > 90) | (df['writing score'] > 90)]
```

```python
print(f"no. of students who have scored more than 90 in writing are {df[(df['writing score']>90)]['gender'].count()}")
print(f"no. of students who have scored more than 90 in reading are {df[(df['reading score']>90)]['gender'].count()}")
print(f"no. of students who have scored more than 90 in math  are {df[(df['math score']>90)]['gender'].count()}")
```

```python
print(f"no. of students who have scored more than 90 in all subjects are {df[(df['math score'] > 90) & (df['reading score'] > 90) & (df['writing score'] > 90)].shape[0]}")
df[(df['math score'] > 90) & (df['reading score'] > 90) & (df['writing score'] > 90)]
```

```python
df2=df[(df['math score'] > 90) & (df['reading score'] > 90) & (df['writing score'] > 90)]
```

```python
print(f"no. of students who have scored less than 70 in all 3 subjects are {df[(df['math score'] < 70) & (df['reading score'] < 70) & (df['writing score'] < 70)].shape[0]}")
df[(df['math score']<70) & (df['reading score']<70) & (df['writing score']<70)]
```

```python
df3=df[(df['math score']<70) & (df['reading score']<70) & (df['writing score']<70)]
```

```python
print(df2["gender"].value_counts())
h = sns.countplot(x="gender",data=df2,palette={"female":"red","male":"blue"})
h.set_title("gender of students who scored more than 90 in all three subjects",y=1.02)
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
result = df.groupby("gender")[numeric_df.columns].mean()
print(result)
```

```python
df.groupby("gender")[numeric_df.columns].mean().plot(kind='pie',y='math score',autopct='%1.0f%%')
df.groupby("gender")[numeric_df.columns].mean().plot(kind='pie',y='reading score',autopct='%1.0f%%')
df.groupby("gender")[numeric_df.columns].mean().plot(kind='pie',y='writing score',autopct='%1.0f%%')
df.groupby("gender")[numeric_df.columns].mean().plot(kind='pie',y='average',autopct='%1.0f%%')
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('math scores')
sns.violinplot(y='math score',data=df,color='orange',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('reading scores')
sns.violinplot(y='reading score',data=df,color='cyan',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('writing scores')
sns.violinplot(y='writing score',data=df,color='green',linewidth=3)
```

```python
fig, ax= plt.subplots(figsize = (10,6))
print(df2["race/ethnicity"].value_counts())
j = sns.countplot(x="race/ethnicity",data=df2.sort_values(by = "race/ethnicity"))
j.set_title("RACE/ETHNICITY OF STUDENTS WHO SCORED MORE THAN 90 IN ALL 3 SUBJECTS", y = 1.02)
plt.show()
```

```python
fig, ax= plt.subplots(figsize = (10,6))
print(df3["race/ethnicity"].value_counts())
j = sns.countplot(x="race/ethnicity",data=df3.sort_values(by = "race/ethnicity"))
j.set_title("RACE/ETHNICITY OF STUDENTS WHO SCORED lESS THAN 70 IN ALL 3 SUBJECTS", y = 1.02)
plt.show()
```

```python
plt.figure(figsize=(13,4))
plt.subplot(1, 3, 1)
plt.title('reading scores')
sns.barplot(x='race/ethnicity',y="reading score",data=df)
plt.subplot(1, 3, 2)
plt.title('writing scores')
sns.barplot(x='race/ethnicity',y="writing score",data=df)
plt.subplot(1, 3, 3)
plt.title('math scores')
sns.barplot(x='race/ethnicity',y="math score",data=df)
```

```python

sns.catplot(x="race/ethnicity",y="average",kind="boxen",data=df.sort_values("race/ethnicity"))
plt.title('race/ethnicity vs average',fontsize=15, fontweight='bold')
plt.show()
```

```python
order = ["master's degree", "bachelor's degree","associate's degree", "some college", "high school","some high school"]
fig, ax= plt.subplots(figsize= (10,6))
sns.set_palette("Oranges")
g= sns.barplot(x="average",y="parental level of education", data=df, order = order)
g.set_title("parental level of education of all students",y=1.02)
plt.show()
```

```python
fig ,ax = plt.subplots(figsize=(11,9))
sns.set_palette("RdBu")
g= sns.barplot(y="average",
              x="parental level of education",
              hue="gender",
              data= df, order = order)
g.set_title("Parental Level Of education of students genderwise", y=1.02)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
print(df3["parental level of education"].value_counts())
n = sns.countplot(x = "parental level of education", data=df3.sort_values(by ="parental level of education"))
n.set_title("PARENTAL LEVEL OF EDUCATION OF THE BOTTOM SCORERS", y=1.02)
plt.show()
```

```python
sns.set_palette("pastel")
plt.figure(figsize=[20,5])
plt.title("effect of the test preparatio course", fontsize=18)
plt.xlabel('xlabel',fontsize=18)
plt.ylabel('ylabel', fontsize=18)
sns.boxplot(data=df, x="test preparation course",y="average")
```

```python
fig, ax = plt.subplots(figsize = (10,6))
sns.set_palette("deep")
print(df3["test preparation course"].value_counts())
j=sns.countplot(x="test preparation course", data=df3.sort_values(by = "test preparation course"))
j.set_title("test preparation course completion of bottom scorers", y=1.02)
plt.show
```

```python
plt.figure(figsize=(13, 4))
green_shades = ["#d0f0c0", "#90ee90", "#32cd32", "#006400"]
plt.subplot(1, 4, 1)
plt.title('Reading Scores')
sns.barplot(x='lunch', y='reading score', data=df, palette=green_shades)
plt.subplot(1, 4, 2)
plt.title('Writing Scores')
sns.barplot(x='lunch', y='writing score', data=df, palette=green_shades)
plt.subplot(1, 4, 3)
plt.title('Math Scores')
sns.barplot(x='lunch', y='math score', data=df, palette=green_shades)
plt.subplot(1, 4, 4)
plt.title('Average Scores')
sns.barplot(x='lunch', y='average', data=df, palette=green_shades)
plt.tight_layout()
plt.show()
```
